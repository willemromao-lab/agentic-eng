## **Saída em streaming**


## **Quando usar streaming**

O streaming permite receber a resposta aos poucos, em vez de esperar o texto completo ao final. Isso melhora a experiência interativa em aplicações que precisam mostrar progresso em tempo real.

Ao analisar o código, note que cada pedaço da saída chega como um fragmento e pode vir acompanhado de metadados sobre sua origem.


```python
for token, metadata in agent.stream(
    {"messages": [HumanMessage(content="Conte tudo sobre Luna City, a capital da Lua")]},
    stream_mode="messages"
):

    # token é um fragmento da mensagem com parte do conteúdo
    # metadata informa qual nó produziu o token
    
    if token.content:  # Verifica se existe conteúdo de fato
        print(token.content, end="", flush=True)  # Exibe o token
```

# **Prompting**

Aqui o foco é mostrar que a qualidade das respostas depende fortemente da forma como instruímos o modelo. Você verá desde prompts simples até estruturas mais controladas, com exemplos e saídas estruturadas.

Como material de estudo, observe que um bom prompt costuma deixar claro **papel**, **objetivo**, **formato esperado** e, quando útil, **exemplos**. Esse padrão reduz ambiguidades e tende a produzir respostas mais consistentes.


## **Prompting básico**


## **Primeiro passo no design de prompts**

Aqui vemos o contraste entre uma pergunta simples e o uso de um `system_prompt` para definir um papel ao modelo. Esse papel orienta o estilo e o tipo de resposta esperada.

Como exercício mental, compare as respostas obtidas com e sem instrução de sistema. Essa diferença é a base do trabalho com prompting.


## **Pergunta direta ao agent**

Neste primeiro exemplo, enviamos uma pergunta simples para estabelecer uma linha de base. A partir dela, ficará mais fácil perceber o impacto das instruções adicionais nos próximos blocos.


```python
question = HumanMessage(content="Qual é a capital da lua?")

response = agent.invoke(
    {"messages": [question]}
)

print(response['messages'][1].content)
```

## **Definindo papel com system prompt**

O `system_prompt` define a persona ou a regra geral de comportamento do agent. Esse mecanismo é um dos pilares do prompting, porque orienta o estilo da resposta desde o início.


```python
system_prompt = "Você é um escritor de ficção científica. Crie uma capital quando o usuário solicitar."

scifi_agent = create_agent(
    model=model,
    system_prompt=system_prompt
    
)

response = scifi_agent.invoke(
    {"messages": [question]}
)

print(response['messages'][1].content)
```

## **Exemplos few-shot**


## **Aprendendo pelo padrão dos exemplos**

A técnica de *few-shot* fornece exemplos dentro do próprio prompt. Com isso, o modelo tende a imitar o formato, o tom e a lógica demonstrados.

Ao estudar, repare que os exemplos não ensinam apenas conteúdo: eles também ensinam **como responder**.


## **Usando exemplos dentro do prompt**

Neste bloco, o prompt já contém demonstrações de entrada e saída. Isso ajuda o modelo a reproduzir o padrão desejado sem precisar de treinamento adicional.


```python
system_prompt = """

Você é um escritor de ficção científica. Crie uma capital espacial quando o usuário solicitar.

Usuário: Qual é a capital de Marte?
Escritor de ficção científica: Marsialis

Usuário: Qual é a capital de Vênus?
Escritor de ficção científica: Venusovia

"""

scifi_agent = create_agent(
    model=model,
    system_prompt=system_prompt
)

response = scifi_agent.invoke(
    {"messages": [question]}
)

print(response['messages'][1].content)
```

## **Prompts estruturados**


## **Controlando o formato da resposta**

Quando especificamos uma estrutura explícita, aumentamos a previsibilidade da saída. Isso é especialmente útil quando a resposta será lida por pessoas ou processada por código.

Observe como o prompt passa a indicar campos esperados. Esse tipo de instrução ajuda a reduzir respostas soltas ou desorganizadas.


## **Forçando uma estrutura de apresentação**

Aqui o foco não é apenas o conteúdo, mas também o formato da resposta. Esse tipo de instrução é útil quando você deseja uniformidade entre diferentes execuções.


```python
system_prompt = """

Você é um escritor de ficção científica. Crie uma capital espacial quando o usuário solicitar.

Mantenha a estrutura abaixo.

Nome: O nome da capital

Localização: Onde ela está localizada

Estilo: 2 a 3 palavras para descrever a atmosfera

Economia: Principais indústrias

"""

scifi_agent = create_agent(
    model=model,
    system_prompt=system_prompt
)

response = scifi_agent.invoke(
    {"messages": [question]}
)

print(response['messages'][1].content)
```

## **Saída estruturada**


## **Da resposta textual para dados organizados**

Além de pedir uma estrutura no texto, podemos definir um esquema com Pydantic para receber uma saída mais organizada. Isso facilita validação, extração de campos e integração com outras partes do sistema.

Este padrão é muito útil quando você precisa transformar a resposta do modelo em dados consumíveis por software.


## **Definindo um esquema de dados**

Com Pydantic, descrevemos exatamente os campos esperados. Em vez de apenas ler um texto, passamos a trabalhar com um objeto estruturado.


```python
from pydantic import BaseModel

class CapitalInfo(BaseModel):
    name: str
    location: str
    vibe: str
    economy: str

agent = create_agent(
    model=model,
    system_prompt="Você é um escritor de ficção científica. Crie uma capital quando o usuário solicitar.",
    response_format=CapitalInfo
)

question = HumanMessage(content="Qual é a capital da Lua?")

response = agent.invoke(
    {"messages": [question]}
)

response["structured_response"]
```

```python
response["structured_response"].name
```

```python
capital_info = response["structured_response"]

capital_name = capital_info.name
capital_location = capital_info.location

print(f"{capital_name} is a city located at {capital_location}")
```

# **Tools**

Nesta seção, você verá como transformar funções Python em **tools** utilizáveis por um agent. Esse é um passo importante porque permite que o modelo execute ações externas em vez de depender apenas do conhecimento interno.

Ao estudar os exemplos, note que a tool precisa ter um propósito bem definido, assinatura simples e descrição clara. Essas informações ajudam o agent a decidir quando e como usá-la.


## **Definição de tools**


## **O que é uma tool na prática**

Uma tool é, em essência, uma função que o agent pode chamar quando precisar executar uma ação externa. No exemplo, a função calcula raiz quadrada.

Ao estudar esta parte, observe a assinatura da função e sua descrição. Esses elementos ajudam o agent a compreender a utilidade da tool.


## **Criando uma função utilizável pelo agent**

A função abaixo é transformada em tool por meio do decorator `@tool`. Esse decorator expõe a função de maneira compreensível para o ecossistema do agent.


```python
from langchain.tools import tool

@tool
def square_root(x: float) -> float:
    """Calcular a raiz quadrada de um número"""
    return x ** 0.5
```

```python
@tool("square_root")
def tool1(x: float) -> float:
    """Calcular a raiz quadrada de um número"""
    return x ** 0.5
```

```python
@tool("square_root", description="Calcular a raiz quadrada de um número")
def tool1(x: float) -> float:
    return x ** 0.5
```

```python
tool1.invoke({"x": 467})
```

## **Adicionando tools ao agent**


## **Ligando capacidade de ação ao agent**

Depois de definir a tool, precisamos passá-la ao agent. A partir daí, o modelo pode decidir utilizá-la para resolver determinadas solicitações.

Nos exemplos abaixo, vale inspecionar não só a resposta final, mas também as mensagens intermediárias e os `tool_calls`, que mostram quando a tool foi acionada.


## **Disponibilizando a tool para o agent**

Depois de criada, a tool é passada ao `create_agent`. A partir desse momento, o agent pode decidir chamá-la quando a tarefa exigir cálculo.


```python
from langchain.agents import create_agent

agent = create_agent(
    model=model,
    tools=[tool1],
    system_prompt="Você é um especialista em aritmética. Use suas tools para calcular a raiz quadrada e o quadrado de qualquer número."
)
```

```python
question = HumanMessage(content="Qual é a raiz quadrada de 467?")

response = agent.invoke(
    {"messages": [question]}
)

print(response['messages'][-1].content)
```

```python
pprint(response['messages'])
```

```python
print(response["messages"][1].tool_calls)
```

# **Web Search**

Aqui o objetivo é comparar dois cenários: um agent sem acesso à web e outro capaz de consultar informações externas. Isso evidencia por que tools são úteis quando o modelo precisa lidar com fatos recentes ou verificar dados fora do treinamento.

Como ponto de estudo, perceba que o agent continua sendo o mesmo conceito geral; o que muda é a sua capacidade operacional ao receber uma tool de busca.


## **Sem busca na web**


## **Limitação sem acesso externo**

Neste primeiro cenário, o agent responde apenas com base em seu treinamento e contexto interno. Isso funciona em muitos casos, mas pode ser insuficiente para informações recentes ou muito específicas.

A ideia aqui é criar uma referência para comparar com o comportamento após adicionar a tool de busca.


```python
question = HumanMessage(content="Quão atualizados são os seus conhecimentos de treinamento?")

response = agent.invoke(
    {"messages": [question]}
)
```

```python
print(response['messages'][-1].content)
```

## **Adicionando uma tool de busca na web**


## **Expandindo o agent com acesso à web**

Agora o agent recebe uma tool capaz de consultar a web. Isso amplia sua utilidade em tarefas que exigem atualização ou verificação externa.

Ao estudar o exemplo, observe a separação entre a definição da tool, a criação do agent e a pergunta enviada em seguida.


## **Implementando a busca como tool**

A função abaixo encapsula a busca na web usando o cliente Tavily. O importante, do ponto de vista didático, é notar que o agent não fala com a web sozinho: ele faz isso por meio de uma tool registrada.


```python
from langchain.tools import tool
from typing import Dict, Any
from tavily import TavilyClient

tavily_client = TavilyClient()

@tool
def web_search(query: str) -> Dict[str, Any]:

    """Buscar informações na web"""

    return tavily_client.search(query)

web_search.invoke("Quem é o atual prefeito de Natal RN?")
```

```python
agent = create_agent(
    model=model,
    tools=[web_search]
)

question = HumanMessage(content="Quem é o atual prefeito de Natal RN?")

response = agent.invoke(
    {"messages": [question]}
)
```

```python
pprint(response['messages'])
```

# **Memory**

Nesta parte, o objetivo é mostrar como a memória muda o comportamento do agent em conversas de múltiplos turnos. Sem memória, cada interação tende a ser tratada de forma isolada; com memória, o histórico pode ser reaproveitado.

Ao estudar, observe o papel do `thread_id`: ele identifica a conversa e permite que o agent recupere o contexto salvo anteriormente.


## **Sem memória**


## **Conversas sem continuidade**

Aqui vemos um agent sem mecanismo de memória. Isso significa que cada chamada é tratada isoladamente, a menos que todo o histórico seja reenviado manualmente.

O objetivo do exemplo é mostrar por que uma informação dita em uma etapa pode não ser lembrada na seguinte.


```python
agent = create_agent(
    model
)
```

```python
question = HumanMessage(content="Olá, meu nome é Willem e minha cor favorita é azul")

response = agent.invoke(
    {"messages": [question]} 
)
```

```python
pprint(response)
```

```python
question = HumanMessage(content="Qual é a minha cor favorita?")

response = agent.invoke(
    {"messages": [question]} 
)

pprint(response)
```

## **Com memória**


## **Mantendo contexto entre interações**

Com um `checkpointer`, o agent passa a preservar informações associadas a uma conversa. Assim, uma pergunta posterior pode reutilizar o que foi informado antes.

Repare que a configuração inclui um identificador de thread. Ele é essencial para que o histórico correto seja recuperado.


## **Adicionando armazenamento de contexto**

Neste bloco, o `InMemorySaver` funciona como um mecanismo simples de persistência do estado da conversa. Ele permite recuperar informações previamente trocadas dentro da mesma thread.


```python
from langgraph.checkpoint.memory import InMemorySaver  


agent = create_agent(
    model,
    checkpointer=InMemorySaver(),  
)
```

```python
question = HumanMessage(content="Olá, meu nome é Willem e minha cor favorita é azul")
config = {"configurable": {"thread_id": "1"}}

response = agent.invoke(
    {"messages": [question]},
    config,  
)
```

```python
pprint(response)
```

```python
question = HumanMessage(content="Qual é a minha cor favorita?")

response = agent.invoke(
    {"messages": [question]},
    config,  
)

pprint(response)
```

# **Multimodal Messages**

Esta seção apresenta entradas multimodais: texto, imagem e áudio. Em vez de enviar apenas uma string, podemos construir mensagens com diferentes tipos de conteúdo, o que amplia bastante as aplicações do modelo.

O principal conceito aqui é que a mensagem passa a ser uma estrutura composta. Cada item informa ao modelo **que tipo de dado está sendo enviado** e **qual conteúdo deve ser interpretado**.


## **Entrada de texto**


## **Mensagens multimodais também podem começar pelo texto**

Mesmo quando usamos apenas texto, já podemos montar a mensagem em um formato multimodal, com listas de blocos contendo `type` e `text`.

Esse padrão é útil porque prepara o terreno para combinar diferentes mídias na mesma mensagem.


## **Criando um agent para receber mensagens multimodais**

O agent continua sendo criado normalmente. O que muda mais adiante é a estrutura do conteúdo enviado dentro da mensagem.


## **Entrada de imagem**


## **Enviando imagem para o modelo**

Nesta subseção, a imagem é carregada, convertida para base64 e enviada como parte de uma mensagem multimodal. Esse processo é comum em integrações com APIs que recebem o conteúdo da imagem em formato textual codificado.

Ao estudar, acompanhe o fluxo completo: upload, leitura do conteúdo, conversão para bytes, codificação em base64 e envio ao modelo.


## **Preparando upload de imagem no notebook**

Antes de enviar a imagem ao modelo, precisamos disponibilizá-la dentro do ambiente do notebook. O widget abaixo permite selecionar um arquivo `.png` manualmente.


```python
from ipywidgets import FileUpload
from IPython.display import display

uploader = FileUpload(accept='.png', multiple=False)
display(uploader)
```

```python
print(uploader.value)
```

## **Convertendo a imagem para base64**

Muitas APIs recebem arquivos binários codificados em base64. Por isso, o passo de conversão é comum em exemplos multimodais.


```python
import base64

# Obtém o primeiro dicionário de arquivo enviado (e o único)
uploaded_file = uploader.value[0]

# Isto é um memoryview
content_mv = uploaded_file["content"]

# Converte memoryview para bytes
img_bytes = bytes(content_mv)  # or content_mv.tobytes()

# Agora codifica em base64
img_b64 = base64.b64encode(img_bytes).decode("utf-8")
```

```python
multimodal_question = HumanMessage(content=[
    {"type": "text", "text": "Fale sobre esta imagem"},
    {"type": "image", "base64": img_b64, "mime_type": "image/png"}
])

response = agent.invoke(
    {"messages": [multimodal_question]}
)

print(response['messages'][-1].content)
```

## **Entrada de áudio**


## **Enviando áudio para o modelo**

Aqui o exemplo grava um áudio, converte o arquivo para bytes e o envia em base64. O processo é semelhante ao da imagem, mas usando um tipo de mídia diferente.

Observe que também muda o modelo utilizado, já que nem todo modelo oferece suporte às mesmas modalidades de entrada.


## **Capturando áudio no próprio notebook**

Neste exemplo, o áudio é gravado diretamente no ambiente local. Depois disso, ele é transformado em bytes e enviado para o modelo.


```python
import sounddevice as sd
from scipy.io.wavfile import write
import base64
import io
import time
from tqdm import tqdm

# Configurações de gravação
duration = 5  # segundos
sample_rate = 44100

print("Gravando...")
audio = sd.rec(int(duration * sample_rate), samplerate=sample_rate, channels=1)
# Barra de progresso durante a gravação
for _ in tqdm(range(duration * 10)):   # atualiza 10× por segundo
    time.sleep(0.1)
sd.wait()
print("Concluído.")

# Write WAV to an in-memory buffer
buf = io.BytesIO()
write(buf, sample_rate, audio)
wav_bytes = buf.getvalue()

aud_b64 = base64.b64encode(wav_bytes).decode("utf-8")
```

```python
agent = create_agent(
    model='gpt-4o-audio-preview',
)

multimodal_question = HumanMessage(content=[
    {"type": "text", "text": "Fale sobre este arquivo de áudio"},
    {"type": "audio", "base64": aud_b64, "mime_type": "audio/wav"}
])

response = agent.invoke(
    {"messages": [multimodal_question]}
)

print(response['messages'][-1].content)
```

# **Personal Chef**

Este tema reúne vários conceitos anteriores em um exemplo mais aplicado: um agent com prompt de sistema, memory e tool de busca na web. A proposta é simular um assistente culinário capaz de sugerir receitas a partir dos ingredientes disponíveis.

Como material de estudo, vale observar como diferentes componentes se combinam: o prompt define o comportamento, a tool amplia o acesso a informações e a memória mantém a continuidade da conversa.


## **Criando a tool de busca para receitas**

Como as sugestões culinárias podem depender de receitas e referências externas, usamos uma tool de busca na web. Esse recurso complementa o conhecimento do modelo.


```python
@tool
def web_search(query: str) -> Dict[str, Any]:

    """Buscar informações na web"""

    return tavily_client.search(query)
```

## **Definindo o comportamento do chef pessoal**

O `system_prompt` estabelece o papel do agent, o tipo de informação que ele deve buscar e o que deve entregar ao usuário.


```python
system_prompt = """

Você é um chef pessoal. O usuário fornecerá uma lista de ingredientes que sobraram em sua casa.

Usando a tool de busca na web, pesquise receitas que possam ser feitas com os ingredientes disponíveis.

Retorne sugestões de receitas e, se solicitado, também as instruções de preparo.

"""
```

## **Montando um agent completo**

Aqui combinamos os elementos estudados anteriormente: modelo, tool, prompt de sistema e memória. Esse é um bom exemplo de composição de recursos em LangChain.


```python
agent = create_agent(
    model=model,
    tools=[web_search],
    system_prompt=system_prompt,
    checkpointer=InMemorySaver()
)
```

```python
config = {"configurable": {"thread_id": "1"}}

response = agent.invoke(
    {"messages": [HumanMessage(content="Tenho frango e arroz sobrando. O que posso preparar?")]},
    config
)

print(response['messages'][-1].content)
```

```python
pprint(response)
```
