```python
import os
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv()

endpoint = "https://senac-labs.cognitiveservices.azure.com/openai/v1/"
api_key = os.getenv("AZURE_OPENAI_API_KEY")

client = OpenAI(
    base_url=f"{endpoint}",
    api_key=api_key
)

response = client.responses.create(
    model="gpt-5-nano",
    input=[{
        "role": "user",
        "content": [
            {"type": "input_text", "text": "O que tem nessa imagem?"},
            {
                "type": "input_image",
                "image_url": "https://www.neilsahota.com/wp-content/uploads/2022/09/What-is-AI-how-does-it-work.jpg",
            },
        ],
    }],
)

print(response.output_text)
```
